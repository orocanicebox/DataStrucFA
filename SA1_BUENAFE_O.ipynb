```python
import random
import time as time
import numpy as np
import matplotlib.pyplot as plt
```

```python
def measure_time(size):
   
    data=[random.randint(0, 1000000) for _ in range(size)]

    start= time.time()
    sorted_data= sorted(data) 
    end= time.time()
    
    elapsed= end - start
    
    return elapsed

sizes=[10, 100, 10000, 1000000, 10000000]
elapsed_times= []


for size in sizes:
    elapsed = measure_time(size)
    elapsed_times.append(elapsed)
    print(f'Size: {size}, Elapsed time: {elapsed:.5f} seconds')


plt.figure(figsize=(10, 10))
plt.plot(sizes, elapsed_times, marker='o', label='Elapsed Times')
plt.xscale('log') 
plt.yscale('log') 
plt.title("Size vs. Time")
plt.xlabel("Size")
plt.ylabel("Elapsed time in seconds")
plt.grid(True)

x = np.array(sizes)
y = x * np.log2(x) / max(x * np.log2(x)) * max(elapsed_times)  
plt.plot(x, y, label='O(n log n)', linestyle='-', color='lightgreen')

plt.legend()
plt.show()
```
